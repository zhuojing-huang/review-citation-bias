```python
import openreview
```

```python
client = openreview.Client(baseurl='https://api.openreview.net')

submissions = client.get_all_notes(invitation="ICLR.cc/2023/Conference/-/Blind_Submission", details='replies')
```

```python
print(submissions[0])
```

```python
reviews = []
for submission in submissions:
    reviews = reviews + [reply for reply in submission.details["replies"] if reply["invitation"].endswith("Official_Review")]
```

```python
len(reviews)
```

```python
reviews[0]
```

```python
import json

# serialize non-serializable Note objects
def custom_serializer(obj):
    # convert Note to a dictionary
    if hasattr(obj, '__dict__'):
        return obj.__dict__
    # in case of unstandard type, convert to string as a fallback
    return str(obj)

filepath = "/content/drive/MyDrive/Thesis/ICLR2023.json"

with open(filepath, 'w') as json_file:
    json.dump(reviews, json_file, indent=4, default=custom_serializer)

print(f"Data saved to {filepath}")
```
